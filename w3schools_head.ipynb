```python
#first the url of the web
url = "https://www.w3schools.com/html/default.asp"
```

```python
from requests import get
from bs4 import BeautifulSoup
```

```python
#taking response from the web
response = get(url)
response
```

```python
#response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
head = soup.find_all("h1")
len(head)
```

```python
head[0].text
```

```python
head1 = soup.find_all("h2")
k = len(head1)
```
